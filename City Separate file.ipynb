```python
import json
import os
import numpy as np
import pandas as pd
from tqdm import tqdm
```

```python
def dataprep(resname):
    data = {"review_id":[],"business_id":[],"text":[],"stars":[]}
    with open('../input/yelp-dataset/yelp_academic_dataset_review.json') as f:   
        for line in tqdm(f):
            review = json.loads(line)
            data['review_id'].append(review['review_id'])
            data['business_id'].append(review['business_id'])
            data['text'].append(review['text'])
            data['stars'].append(review['stars'])
    df = pd.DataFrame(data)
    data1 = {"name":[],"business_id":[],"city":[],"categories":[]}
    with open('../input/yelp-dataset/yelp_academic_dataset_business.json') as f:
        for line in tqdm(f):
            business = json.loads(line)
            data1['name'].append(business['name'])
            data1['business_id'].append(business['business_id'])
            data1['city'].append(business['city'])
            data1['categories'].append(business['categories'])
    df1 = pd.DataFrame(data1)
    df.dropna(inplace=True)
    df1.dropna(inplace=True)
    restaurants_business = df1[df1.categories.str.contains("Restaurants")]
    Toronto_rest = restaurants_business[restaurants_business.city == "Toronto"]
    finaldf = Toronto_rest.merge(df,on="business_id",how="inner")
    final_reviews = finaldf[finaldf.name==resname]
    review = final_reviews.reset_index().text
    data = pd.DataFrame(review)
return data1,data
```

```python
data = {"review_id":[],"business_id":[],"text":[],"stars":[],"user_id":[]}
with open('data/yelp_academic_dataset_review.json') as f:   
    for line in tqdm(f):
        review = json.loads(line)
        data['review_id'].append(review['review_id'])
        data['business_id'].append(review['business_id'])
        data['user_id'].append(review['user_id'])
        data['text'].append(review['text'])
        data['stars'].append(review['stars'])
df = pd.DataFrame(data)
```
